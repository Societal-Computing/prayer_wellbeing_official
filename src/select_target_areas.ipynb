### Filter based on expected reach from FB ads

```python
raw = pd.read_csv('../input/areas_dominant_religion_with_nonzero_reach.csv')
raw.drop(columns = ['Unnamed: 0'], inplace = True)
```

```python
raw.head(1)
```

##### Extract country names

```python
raw['iso2_cc'] = raw['geoname_adm_code'].str.extract(r'\|([A-Za-z]+)\|')
```

##### Filter target locations

Only precision code 1

```python
df = raw.query("precision_code == 1").copy()
```

For every country, just keep the location with the highest estimated reach

```python
selected = (df
            .loc[df.groupby(['dominant', 'iso2_cc', 'place_name'])['Reach'].idxmax(), 
                 ['iso2_cc', 'place_name', 'dominant', 'latitude', 'longitude', 'Reach']]
            .drop_duplicates(subset = ['iso2_cc', 'dominant'])
            .sort_values(by = ['dominant', 'Reach'], ascending = False)
            .groupby('dominant')
            .head(5)
            .reset_index(drop = True))
```
